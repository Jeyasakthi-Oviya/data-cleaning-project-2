```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_purchase=pd.read_csv('purchases.csv')
data_purchase.head()
```

```python
data_purchase.tail()
```

```python
data_purchase.info()
```

```python
data_purchase.isnull().sum()
```

```python
data_purchase=data_purchase.drop(['CustomerID','PurchaseDate'],axis=1)
```

```python
data_purchase
```

```python
def change_quantity(Quantity):
    if str(Quantity).isdigit():
        return int(Quantity)
    else:
        return None
    
```

```python
data_purchase['Quantity']=data_purchase['Quantity'].apply(change_quantity)
```

```python
data_purchase
```

```python
data_purchase['Quantity']=data_purchase['Quantity'].fillna(data_purchase['Quantity'].mode()[0]).astype(int)
```

```python
data_purchase
```

```python
def price_change(Price):
    if str(Price).isdigit():
        return int(Price)
    else:
        return None
```

```python
data_purchase['Price']=data_purchase['Price'].apply(price_change)
```

```python
data_purchase
```

```python
data_purchase['Price']=data_purchase['Price'].fillna(data_purchase['Price'].mode()[0]).astype(int)
```

```python
data_purchase
```

```python
data_purchase.isnull().sum()
```

```python
data_purchase['Region']=data_purchase['Region'].str.strip().str.lower()
```

```python
data_purchase
```

```python
data_purchase['Region']=data_purchase['Region'].str.strip().str.capitalize()
```

```python
data_purchase
```

```python
data_purchase
```

```python
data_purchase['Region']=data_purchase['Region'].fillna(data_purchase['Region'].mode()[0])
```

```python
data_purchase
```

```python
data_purchase.isnull().sum()
```

```python
data_purchase['Product']=data_purchase['Product'].str.replace(r'[^0-9]','',regex=True)
data_purchase['Product']='Product'+ data_purchase['Product']
```

```python
data_purchase
```

```python
data_purchase.isnull().sum()
```

```python
sns.histplot(data_purchase['Region'],bins=5)
```

```python
sns.histplot(data_purchase['Price'],bins=5)
```

```python
sns.barplot(x='Region',y='Price',data=data_purchase)
plt.title("Price Vs Region")
plt.show()
```
